## 2.3 JNF: The Case of the $2 \times 2$ Matrix 

Suppose that $A\in\mathbb{R}^{2\times 2}$, then the first step to computing the JNF is to compute the characteristic polynomial, $\chi_A(t)$. There are two possible cases:

<br>

 **Case 1:**

 $$
 \chi_A(t) = (t-\lambda_1)(t-\lambda_2)
 $$

 with $\lambda_1\neq\lambda_2$. 
 
 $A$ is diagonalisable, *i.e.* $\displaystyle  J = \begin{pmatrix}
 \lambda_1 &0\\0&\lambda_2
 \end{pmatrix}$.

 <br>

**Case 2:**

$$
 \chi_A(t) = (t-\lambda)^2.
$$

To begin, calculate the $\lambda$-eigenspace. 

If the GM = 2 then the matrix is diagonalisable, *i.e.* $\displaystyle J = \begin{pmatrix}
\lambda &0\\0&\lambda
\end{pmatrix}$.

Otherwise, we consider the generalised eigenspace $\displaystyle (A-\lambda I)^2$. 

By the C-H we have that

$$
 (A-\lambda I)^2 = 0 \implies (A-\lambda I)^2v = 0 \quad \forall v\in\mathbb{R}^2,
$$

or in other words, the generalised $\lambda$-eigenspace is just $\mathbb{R}^2$. 

In this latter case we have that $\displaystyle J = \begin{pmatrix}
\lambda &1\\0&\lambda
\end{pmatrix}$.

<br>

We need to find a basis consisting of eigenvectors and generalised eigenvectors. In the 2 by 2 case this is quite straightforward:

 1. Choose a vector $\mathbf{v}$ which does not belong to the $\lambda$-eigenspace, \emph{i.e.} is not a multiple of the eigenvector corresponding to $\lambda$.
 2. Define $\displaystyle \mathbf{w} = (A-\lambda I)\mathbf{v}$
 $\mathbf{w}$ is an eigenvector. \quad \textcolor{red}{(Why?)}
 3. Construct $\displaystyle P = (\mathbf{w}, \mathbf{v})$.
 4. Check $P^{-1}AP = J$.

 <br>

**Example 2.3.1** Let $\displaystyle \begin{pmatrix}
4&1\\-1&6
\end{pmatrix}$, then 

$$
\chi_A(t) &= \left|\begin{array}{cc}
4-t&1\\-1&6-t
\end{array}\right|\\
&= (4-t)(6-t)+1\\
&=(t-5)^2.
$$

The $\lambda$-eigenspace of $A$ is given by

$$
\left\{\alpha\begin{pmatrix} 1\\ 1\end{pmatrix} : \alpha\in\mathbb{R}\right\}
$$

and so GM=1.

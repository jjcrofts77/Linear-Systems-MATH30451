# 2.3 JNF: The Case of the $3 \times 3$ Matrix 

As with any such problem we start by calculating the characteristic equation, and then find the eigenvalues by looking for roots of the characteristic equation (in an exam question, you should always be able to spot one root and then factorise from there, so start by checking if $\pm 1, \pm 2$ are roots). There are three main cases, some of which split up into sub cases, but all  follow the same general method: (i) three distinct eigenvalues, (ii) two distinct eigenvalues, i.e., the characteristic equation has one repeated root, and one other root, (iii) one eigenvalue, i.e., the characteristic equation has one thrice repeated root.

<br>

````{margin}
```{important}
In Case (i) the $\lambda_i$ are distinct
```
````
>**Case (i)**  $\mathbf{\displaystyle\chi_A(t) = (t-\lambda_1)(t-\lambda_2)(t-\lambda_3)}$

- Find eigenvectors $v_1, v_2$ and $v_3$ such that
$\displaystyle Av_i=\lambda_iv_i$ for $i=1,2,3$. 

- Define the matrix $\displaystyle P=[v_1, v_2, v_3]$ and then

$$
 P^{-1}AP =
\left(\begin{array}{ccc}\lambda_1&0&0\\0&\lambda_2&0\\0&0&\lambda_3\end{array}
\right).
$$

<br>

>**Case (ii)** $\mathbf{\displaystyle \chi_A(t) = (t-\lambda_1)(t-\lambda_2)^2:}$

- Firstly, find an eigenvector $v_1$ such that $\displaystyle Av_1=\lambda_1v_1$. 

- Next find eigenvectors  for $\lambda_2$, i.e., find a basis of $\displaystyle\ker{(A-\lambda_2I)}$ -- this splits into two sub cases, where $\displaystyle\dim(\ker(A-\lambda_2I))=2$ and where $\dim(\ker(A-\lambda_2I))=1$. 
<br><br>

 Sub-case (a)
: $\mathbf{\displaystyle\dim(\ker(A-\lambda_2I))=2}$ 
 - Pick two
 linearly independent vectors from the $\displaystyle\ker(A-\lambda_2I)$, say  $v_2$ and 
$v_3$,  and form the matrix $\displaystyle P=[v_1, v_2, v_3]$ and then 

$$
P^{-1}AP=\left(\begin{array}{ccc}\lambda_1&0&0\\0&\lambda_2&0\\0&0&\lambda_2
\end{array}\right).
$$
<br>

 Sub-case (b)
: $\mathbf{\displaystyle\dim(\ker(A-\lambda_2I))=1}$ 
 - Let $v_2$ be in $\displaystyle\ker(A-\lambda_2I)$. 
 - To find a third vector $v_3$ we consider the *generalised eigenspace* which is given by $\displaystyle\ker\left[(A-\lambda_2I)^2\right]$, and is $2$-dimensional since $\mathbb{R}^3$ is a direct sum of generalised eigenspaces. So pick a vector $v_3$ which is in
$\displaystyle\ker{\left[(A-\lambda_2I)^2\right]}$, so that $v_1, v_2$ and $v_3$ form a
basis  of $\mathbb{R}^3$. 
 - Modify the basis by replacing $v_2$ by $\displaystyle v'_2=(A-\lambda_2I)v_3$. Then since $\displaystyle v_3\in\ker\left[(A-\lambda_2I)^2\right]$ this means that $\displaystyle v'_2\in\ker(A-\lambda_2I)$. 
 - Forming the matrix $\displaystyle P=[v_1,v'_2,v_3]$ we obtain

$$
 P^{-1}AP = \left(\begin{array}{ccc}\lambda_1&0&0\\0&\lambda_2&1\\0&0&\lambda_2
\end{array}\right).
$$

<br>

>**Case (iii)** $\mathbf{\displaystyle\chi_A(t)=(t-\lambda)^3}:$ 

- As always, start by finding the eigenvectors which satisfy $\displaystyle (A-\lambda I)v=0$. There are sub cases according to whether the dimension of $\displaystyle\ker(A-\lambda I)$ is $1, 2$ or $3$.
<br><br>

 Sub-case (a)
: $\mathbf{\displaystyle\dim{(\ker(A-\lambda I))}=3:}$ 
 - This means that $\displaystyle Av=\lambda v$ for all vectors in $\mathbb{R}^3$, which means that $\displaystyle A=\lambda I$, and $A$ is already in Jordan normal form. 

 <br>

 Sub-case (b)
: $\mathbf{\displaystyle\dim{(\ker(A-\lambda I))}=2:}$ 
 
 - Pick linearly independent vectors $v_1$ and $v_2$ which span $\displaystyle\ker(A-\lambda I)$. 

 - Proposition 2.3 implies that $\displaystyle\ker\left[(A-\lambda I)^2\right]=\mathbb{R}^3$, so pick $v_3$ which is in $\displaystyle\ker\left[(A-\lambda I)^2\right]$ but not in $\displaystyle\ker(A-\lambda I)$ so that $v_1, v_2$ and $v_3$ form a basis of $\mathbb{R}^3$. 

 - Now modify this basis by replacing $v_1$ with $\displaystyle v'_1 = (A-\lambda I)v_3$. As in previous cases, $\displaystyle v_3\in\ker\left[(A-\lambda I)^2\right]$ implies that $\displaystyle v'_1\in\ker(A-\lambda I)$, so $v'_1, v_2$ and $v_3$ still form a basis of $\mathbb{R}^3$.

 - Now form the matrix $\displaystyle P=[v'_1, v_3, v_2]$, and note the order that the vectors are taken. Then

 $$
P^{-1}AP=\left(\begin{array}{ccc}\lambda&1&0\\0&\lambda&0\\0&0&\lambda
\end{array}\right).
$$
<br>

 Sub-case (c)
: $\mathbf{\displaystyle \dim{\ker(A-\lambda I)}=1:}$. 

 - Pick vector $v_1$ which spans $\displaystyle\ker(A-\lambda I)$. 
 
 - Now Proposition 2.3 says that 

 $$\displaystyle\ker(A-\lambda I)\subset\ker\left[(A-\lambda I)^2\right]\subset\ker\left[(A-\lambda I)^3\right]$$

  - Since we cannot have $\displaystyle\dim(\ker\left[(A-\lambda I)^2\right])=3$ in this case, we extend $v_1$ to a basis of $\displaystyle\ker\left[(A-\lambda I)^2\right]$ given by $v_1, v_2$ (i.e., find such a $v_2$). 
 
 - Then extend again to get a basis of $\displaystyle\ker\left[(A-\lambda I)^3\right]$ (which is $\mathbb{R}^3$ by
Cayley-Hamilton) given by $v_1, v_2$ and $v_3$.

 - Now, modify this basis by replacing $v_2$ by $\displaystyle v'_2=(A-\lambda I)v_3$ and by replacing $v_1$ with 
 
 $$\displaystyle v'_1=(A-\lambda I)v'_2=(A-\lambda I)^2v_3.$$ 
 
 - It follows that $v'_1, v'_2$ and $v_3$ gives a basis of $\mathbb{R}^3$. 
 
 - Form the matrix $\displaystyle P=[v'_1, v'_2, v_3]$ and then 

$$
  P^{-1}AP = \left(\begin{array}{ccc}\lambda&1&0\\0&\lambda&1\\0&0&\lambda\end{array}\right).
  $$

  <br>

   Let us now work through a few examples to help clarify the above.

  <br>

**Example 2.3.1**  Let

$$
A = \left(\begin{array}{rc}4&1\\-1&6\end{array}\right).
$$

<br>

Then the characteristic equation is given by $\displaystyle\chi_A(t)=(4-t)(6-t)+1=(t-5)^2$, so that 5 is the only eigenvalue. 

Now, since $A$ is not a multiple of the identity matrix it follows that the Jordan normal form must be given by

$$
\left[\begin{array}{cc}5&1\\0&5\end{array}\right],
$$

so now we just have to find a base change matrix to get to it. So, find the eigenvector:

$$
\left(\begin{array}{rr}4&1\\-1&6\end{array}\right)\left(\begin{array}{c}x\\y    \end{array}\right) = \left(\begin{array}{c}5x\\5y\end{array}\right).
$$

Thus $\displaystyle 4x+y=5x$ and $\displaystyle -x+6y=5y$ which reduce to $y=x$, so define $\displaystyle v_1=[1, 1]^T$. 

Now extend to a basis of $\mathbb{R}^2$ by, for example, $\displaystyle v_2=[1, 0]^T$. 

Next modify the basis by taking $\displaystyle v'_1=(A-5I)v_2$ which if you work this out is $\displaystyle [-1, -1]^T$. 
 
From which we obtain the matrix (*similarity transformation*) and its inverse:

$$
P=\left(v'_1, v_2\right)=\left(\begin{array}{cc}-1&1\\-1&0\end{array}\right) \text{ and } P^{-1}=\left(\begin{array}{cc}0&-1\\1&-1\end{array}\right).
$$

It is left as an exercise to check that 

$$
P^{-1}AP = \left(\begin{array}{cc}5&1\\0&5\end{array}\right).
$$

<br>

**Example 2.3.2** Let

$$
A=\left(\begin{array}{rrr}-3&-1&1\\-1&-3&1\\-2&-2&0\end{array}\right).
$$

<br>

Working out the characteristic equation gives $\displaystyle\chi_A(t) = (t+2)^3$. So $-2$ is the only eigenvalue. Now find eigenvectors,

<br>

$$
 \left(\begin{array}{rrr}-3&-1&1\\-1&-3&1\\-2&-2&0\end{array}\right)\left(\begin
{array}{c}x\\y\\z\end{array}\right) =
\left(\begin{array}{c}-2x\\-2y\\-2z\end{array}\right). 
$$

<br>

These equations reduce to $\displaystyle -x-y+z=0$, so $\displaystyle\ker(A+2I)$ is $2$-dimensional and we are in case (iii) of the $3\times 3$ section. 

Pick a basis of $\displaystyle\ker(A+2I)$, *e.g.*, $\displaystyle v_1=[1, -1,0]^T$ and $\displaystyle v_2=[0, 1, 1]^T$. 

Now, $\displaystyle\ker\left[(A+2I)^2\right]$ must be $\mathbb{R}^3$, so extend $v_1$ and $v_2$ to a basis of $\mathbb{R}^3$ by, for example, $\displaystyle v_3=[1, 0, 0]^T$. 

Next modify the basis by replacing $v_1$ with $\displaystyle v'_1=(A+2I)v_3=[-1,-1,-2]^T$. 

Form the matrix

$$
P=\left(v'_1,v_2,v_3\right)=\left(\begin{array}{rrr}-1&1&0\\-1&0&1\\-2&0&1\end{array}
\right).
$$

We can now check that 

$$
P^{-1}=\left(\begin{array}{rrr}0&1&-1\\1&1&-1\\0&2&-1\end{array}\right)
\text{ and that }
P^{-1}AP=\left(\begin{array}{rrr}-2&1&0\\0&-2&0\\0&0&-2\end{array}\right),
$$

as expected.

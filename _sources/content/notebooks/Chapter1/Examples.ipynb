

**Example 1.1.1** Assuming that damping effects are negligible, then for a vibrating system as shown the displacements $x_1, x_2$ of the masses $m_1, m_2$ from the equilibrium position are given by the system of equations

$$ m_1\frac{\mathrm{d}^2x_1}{\mathrm{d}t^2} &= -k_1x_1-k_3(x_1-x_2),\\ m_2\frac{\mathrm{d}^2x_2}{\mathrm{d}t^2} &= -k_2x_2-k_3(x_2-x_1). $$

That is

$$
 \frac{\mathrm{d}^2}{\mathrm{d}t^2}\left(\begin{array}{c}x_1\\x_2\end{array}\right) = 
 \left(\begin{array}{cc}-\frac{k_1+k_3}{m_1} & \frac{k_3}{m_1}\\
\frac{k_3}{m_2}&
-\frac{k_2+k_3}{m_2}\end{array}\right)\left(\begin{array}{c}x_1\\x_2\end{array}
\right),
$$

or

$$
 \frac{\mathrm{d}^2\mathbf{x}}{\mathrm{d}t^2} = A\mathbf{x},
$$

where 

$$
\mathbf{x} = \begin{pmatrix} x_1\\ x_2\end{pmatrix} \text{ and } A =
\left(\begin{array}{cc}-\frac{k_1+k_3}{m_1} & \frac{k_3}{m_1}\\
\frac{k_3}{m_2}& -\frac{k_2+k_3}{m_2}\end{array}\right).
$$

Alternatively, the above system of second order differential equations can be replaced by one containing first order ones by writing $\displaystyle\frac{\mathrm{d}x_1}{\mathrm{d}t}=v_1$ and $\displaystyle\frac{\mathrm{d}x_2}{\mathrm{d}t}=v_2$, so that

$$
 m_1\frac{\mathrm{d}v_1}{\mathrm{d}t} &= -k_1x_1-k_3(x_1-x_2),\\
 m_2\frac{\mathrm{d}v_2}{\mathrm{d}t} &= -k_2x_2-k_3(x_2-x_1).
$$

Then

$$
\frac{\mathrm{d}}{\mathrm{d}t}\left(\begin{array}{c}x_1\\x_2\\v_1\\v_2\end{array}\right) = 
\left(\begin{array}{cccc}0&0&1&0\\0&0&0&1\\-\frac{k_1+k_3}{m_1}&\frac{k_3}{m_1}
&0&0\\\frac{k_3}{m_2}&-\frac{k_2+k_3}{m_2}&0&0\end{array}\right)
\begin{pmatrix} x_1\\x_2\\v_1\\v_2\end{pmatrix}.
$$

```{Note} 
It is common practice to replace a single $n$th order
differential equation of the form

$$
\frac{\mathrm{d}^nx}{\mathrm{d}t^n}+a_1\frac{\mathrm{d}^{n-1}x}{\mathrm{d}t^{n-1}}+a_2\frac{\mathrm{d}^{n-2}x}{\mathrm{d}t^{n-2}}
+\cdots +a_{n-1}\frac{\mathrm{d}x}{\mathrm{d}t}+a_nx = 0,
$$

by a system of first order differential equations obtained by writing

$$
\displaystyle x_1=x, x_2=\frac{\mathrm{d}x}{\mathrm{d}t}, x_3=\frac{\mathrm{d}^2x}{\mathrm{d}t^2},\ldots,
x_{n}=\frac{\mathrm{d}^{n-1}x}{\mathrm{d}t^{n-1}}, 
$$

so that 

$$
 \frac{\mathrm{d}x_{n}}{\mathrm{d}t}+a_1x_{n}+a_2x_{n-1}+\cdots+a_{n-1}x_2
+a_nx_1 = 0.
$$

Then

$$
\frac{\mathrm{d}}{\mathrm{d}t}\left(\begin{array}{c}x_1\\x_2\\x_3\\\vdots\\x_{n-1}\\x_{n}
\end{array}\right)
=\left(\begin{array}{cccccc}
0&1&0&\ldots&0&0\\0&0&1&\ldots&0&0\\0&0&0&\ldots&0&0\\
\vdots&\vdots&\vdots&&\vdots&\vdots\\0&0&0&\ldots&0&1\\-a_n&-a_{n-1}&-a_{n-2}
&\ldots&-a_2&-a_1\end{array}\right)
\left(\begin{array}{c}x_1\\x_2\\x_3\\\vdots\\x_{n-1}\\x_{n}\end{array}\right).
$$
```

<br>

**Example 1.1.2** Consider a situation in which three thermally insulated tanks each containing an initial volume $V$ of liquid are connected so that the liquid flows between the tanks at rates as shown and there is a heat flow $u(t)$ into the first tank, N.B. heat flow $=$ liquid flow $\times$ heat per unit volume. Then, if $x_1$, $x_2$ and $x_3$ are the amounts of heat in tanks $1, 2$ and $3$ respectively

$$
 \frac{\mathrm{d}x_1}{\mathrm{d}t} &= u(t)+\frac{2Rx_3}{V} -\frac{2Rx_1}{V},\\
 \frac{\mathrm{d}x_2}{\mathrm{d}t} &= \frac{Rx_1}{V}-\frac{Rx_2}{V},\\
 \frac{\mathrm{d}x_3}{\mathrm{d}t} &= \frac{Rx_1}{V}+\frac{Rx_2}{V}-\frac{2Rx_3}{V}.
$$

That is

$$
\frac{\mathrm{d}}{\mathrm{d}t}\left(\begin{array}{c}x_1\\x_2\\x_3\end{array}\right) =
\frac{R}{V}\left(\begin{array}{ccc}-2&0&2\\1&-1&0\\1&1&-2\end{array}\right)\left
(\begin{array}{c}x_1\\x_2\\x_3\end{array}\right) +
\left(\begin{array}{c}u(t)\\0\\0\end{array}\right),
$$

or

$$
 \frac{\mathrm{d}\mathbf{x}}{\mathrm{d}t} = A\mathbf{x}+\mathbf{f}(t),
$$

where 

$$
\mathbf{x}= \begin{pmatrix} x_1\\x_2\\x_3\end{pmatrix},\quad A=\frac{R}{V}\left(\begin{array}{rrr}-2&0&2\\1&-1&0\\1&1&-2\end{array}\right) \text{ and }\mathbf{f}(t) = \begin{pmatrix} u(t)\\ 0\\ 0\end{pmatrix},
$$

represents a *forcing* or *control* effect on the system.

Now, suppose that, instead of the flow rates being adjusted so that the volume in each of the tanks remains constant, the flow rate from tank $1$ to tank $3$ is $R+r$. Then, the volume of liquid in tank $1$ at time $t$ is $V-rt$ and the volume in tank $3$ is $V+rt$. It follows that 

$$
 \frac{\mathrm{d}x_1}{\mathrm{d}t} &= u(t)+\frac{2Rx_3}{V+rt} -\frac{(2R+r)}{V-rt}x_1,\\
 \frac{\mathrm{d}x_2}{\mathrm{d}t} &= \frac{Rx_1}{V-rt}-\frac{Rx_2}{V},\\
 \frac{\mathrm{d}x_3}{\mathrm{d}t} &= \frac{(R+r)}{V-rt}x_1+\frac{Rx_2}{V}-\frac{2Rx_3}{V+rt},
 $$

or

$$
  \frac{\mathrm{d}\mathbf{x}}{\mathrm{d}t} = A(t)\mathbf{x}+\mathbf{f}(t),
$$

where

$$
A=\left(\begin{array}{rrc}-\frac{2R+r}{V-rt}&0&\frac{2R}{V+rt}\\\frac{R}{V-rt}
&-\frac{R}{V}&0\\\frac{R+r}{V-rt}&\frac{R}{V}&-\frac{2R}{V+rt}\end{array}\right).
$$

In this case the coefficients in the system of differential equations are not constant; however, these differential equations are *still* linear.

<br>

**Example 1.1.3** Consider a situation in which liquid of density $\rho$ is pumped into a tank at rate $w_I$ and then flows into another tank through a valve at rate $w$. If $h_1$ and $h_2$ are the heights of liquids in tanks $1$ and $2$ respectively, then the mass continuity equations for the tanks are

$$
\rho A_1\frac{\mathrm{d}h_1}{\mathrm{d}t} &= w_I-w = w_I-c\sqrt{\rho(h_1-h_2)},\\
\rho A_2\frac{\mathrm{d}h_2}{\mathrm{d}t} &= w = c\sqrt{\rho(h_1-h_2)}.
$$

In this case a non-linear system of equations has been obtained. Non-linear systems cannot be solved by the analytical techniques considered in this course rather numerical techniques are typically considered (See PDEs and/or NADS modules). It may also be possible to approximate a non-linear system by a linear one in some circumstances.

<br>

**Example 1.1.4** Suppose that an animal population can be classified in three ways: young $(1)$, mature $(1-2)$ and old $(2+)$; and that survival rates and birth rates from the three categories are as shown.

```{margin}
![giphy](https://github.com/jjcrofts77/Linear-Systems-MATH30451/blob/main/content/images/Lots_of_rabbits.gif?raw=1)
```

Then if $y_k, m_k$ and $d_k$ are the number in each of the three categories at the start of the $k$th year, it follows that

$$
y_{k+1} &= b_1m_k+b_2d_k,\\
m_{k+1} &= s_1y_k,\\
d_{k+1} &= s_2m_k+s_3d_k.
$$

That is,

$$
\mathbf{p}_{k+1} = A\mathbf{p}_k,
$$

where

$$
\mathbf{p}_k=\begin{pmatrix} y_k\\m_k\\d_k\end{pmatrix} \text{ and }
A = \left(\begin{array}{ccc}0&b_1&b_2\\s_1&0&0\\0&s_2&s_3\end{array}\right).
$$

We can visulise this process using the so-called *state diagram*, which is shown below for our example

```{figure} ../../images/PopDynStateDiag.png
---
height: 350px
name: FigPDSD
---
State diagram for the population model in Example 1.1.4
```
In some circumstances it may be that new animals are introduced into the
population at certain times. In this case the appropriate system would have the form $\displaystyle\mathbf{p}_{k+1} = A\mathbf{p}_k+\mathbf{f}_k$.

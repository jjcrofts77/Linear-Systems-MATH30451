# 3.6 Inhomogeneous Systems of Difference Equations

We shall now turn our attention to systems of first order difference equations of the form

$$
 \mathbf{x}_{k+1} = A\mathbf{x}_k+\mathbf{f}_k
$$ (nonhomodiff)

for which $\displaystyle\mathbf{x}_0$ is given. The solution of such a system may be found inductively as follows 

$$
\text{For } t=0 \quad \mathbf{x}_1 &= A\mathbf{x}_0+\mathbf{f}_0,\\
\text{For } t=1 \quad \mathbf{x}_2 &=A\mathbf{x}_1+\mathbf{f}_1 = A(A\mathbf{x}_0+\mathbf{f}_0)+\mathbf{f}_1,\\
&=A^2\mathbf{x}_0+A\mathbf{f}_0+\mathbf{f}_1,\\
\text{For } t=2 \quad \mathbf{x}_3 &= A\mathbf{x}_2+\mathbf{f}_2 = A(A^2\mathbf{x}_0+A\mathbf{f}_0+\mathbf{f}_1)+\mathbf{f}_2,\\
&=A^3\mathbf{x}_0+A^2\mathbf{f}_0+A\mathbf{f}_1+\mathbf{f}_2,\\
&~~\vdots\\
\mathbf{x}_k &= A^k\mathbf{x}_0 +A^{k-1}\mathbf{f}_0+A^{k-2}\mathbf{f}_{1}+\cdots+\mathbf{f}_{k-1},\\
&=A^k\mathbf{x}_0+\sum_{r=0}^{k-1}{A^r\mathbf{f}_{k-1-r}},\\
&=A^k\mathbf{x}_0+\sum_{r=0}^{k-1}{A^{k-1-r}\mathbf{f}_{r}}.
$$

Alternatively, it could be found by using z-transforms. Taking z-transforms of  {eq}`nonhomodiff`


$$
 z\bar{\mathbf{x}}-z\mathbf{x}_0 = A\bar{\mathbf{x}}+\bar{\mathbf{f}},
$$

which upon rearrangement gives

$$
\left(zI_n-A\right)\bar{\mathbf{x}} = z\mathbf{x}_0+\bar{\mathbf{f}},
$$

or

$$
\bar{\mathbf{x}} = \left(zI_n-A\right)^{-1}\left(z\mathbf{x}_0+\bar{\mathbf{f}}\right).
$$

Now,

$$
\left(zI_n-A\right)^{-1} &= \left[z(I_n-\frac{1}{z}A)\right]^{-1}=\frac{1}{z}(I_n-\frac{1}{z}A)^{-1},\\
&=\frac{1}{z}\sum_{r=0}^\infty{\left(\frac{1}{z}A\right)^r}.
$$

Therefore

$$
 \left(zI_n-A\right)^{-1}z\mathbf{x}_0 = \sum_{r=0}^\infty{\frac{1}{z^r}A^r\mathbf{x}_0},
$$

whose inverse transform is

$$
\sum_{r=0}^\infty{\delta_{kr}A^r\mathbf{x}_0} = A^k\mathbf{x}_0.
$$

Also, 

$$
 \left(zI_n-A\right)^{-1}\bar{\mathbf{f}} = \sum_{r=0}^\infty{\frac{1}{z^{r+1}}A^r\bar{\mathbf{f}}} = \sum_{r=0}^\infty{A^r\frac{1}{z^{r+1}}\bar{\mathbf{f}}},
$$

whose inverse z-transform is given by

$$
 \sum_{r=0}^\infty A^r\mathbf{f}_{k-r-1}u(k-r-1) = \sum_{r=0}^{k-1}A^r\mathbf{f}_{k-r-1}.
$$

<br>

**Example 3.6.1** Consider the difference equation

$$
 x_{k+1} = \frac{1}{3}x_k+1 \quad :\quad x_0=1.
$$

$$
x_k &= \left(\frac{1}{3}\right)^k\cdot 1 + \sum_{r=0}^{k-1}\left(\frac{1}{3}\right)^r\cdot 1,\\
&=\left(\frac{1}{3}\right)^k+\frac{1-\left(\frac{1}{3}\right)^k}{1-\frac{1}{3}},\\
&= \left(\frac{1}{3}\right)^k+\frac{3}{2}\left[1-\left(\frac{1}{3}\right)^k\right]=\frac{3}{2}-\frac{1}{2}\left(\frac{1}{3}\right)^k.
$$

<br>

**Example 3.6.2** Consider 

$$
A = \begin{pmatrix}2&1&1\\-1&0&-1\\3&5&4\end{pmatrix} = HDH^{-1},
$$

where

$$
H=\begin{pmatrix}1&-1&1\\0&1&-1\\-1&-1&2\end{pmatrix},
H^{-1}=\begin{pmatrix}1&1&0\\1&3&1\\1&2&1\end{pmatrix}
~\&~
D=\begin{pmatrix}1&0&0\\0&2&0\\0&0&3\end{pmatrix}.
$$

Here

$$
A^r = HD^rH^{-1} = \begin{pmatrix}1&-1&1\\0&1&-1\\-1&-1&2\end{pmatrix}
\begin{pmatrix}1&0&0\\0&2^r&0\\0&0&3^r\end{pmatrix}\begin{pmatrix}1&1&0\\1&3&1\\1&2&1\end{pmatrix}.
$$

In this case it is probably worthwhile to determine $A^r$ in the light of the example to be considered. 

Now

$$
D^rH^{-1} = \begin{pmatrix}1&0&0\\0&2^r&0\\0&0&3^r\end{pmatrix}\begin{pmatrix}1&1&0\\1&3&1\\1&2&1\end{pmatrix}=\begin{pmatrix}1&1&0\\2^r&3\cdot2^r&2^r\\3^r&2\cdot3^r&3^r\end{pmatrix},
$$

and

$$
A^r &= HD^rH^{-1}=\begin{pmatrix}1&-1&1\\0&1&-1\\-1&-1&2\end{pmatrix}\begin{pmatrix}1&1&0\\2^r&3\cdot2^r&2^r\\3^r&2\cdot3^r&3^r\end{pmatrix},\\
&=\begin{pmatrix}1-2^r+3^r&1-3\cdot2^r+2\cdot3^r&-2^r+3^r\\2^r-3^r&3\cdot2^r-2\cdot3^r&2^r-3^r\\-1-2^r+2\cdot3^r&-1-3\cdot2^r+4\cdot3^r&-2^r+2\cdot3^r\end{pmatrix}.
$$

Suppose now that in the system {eq}`nonhomodiff`, $\displaystyle\mathbf{f}_k=[1,0,0]^T$ and $\displaystyle\mathbf{x}_0=[0,-1,2]^T$. 

Then, using

$$
\mathbf{x}_k = A^k\mathbf{x}_0+\sum_{r=0}^{k-1}A^r\mathbf{f}_{k-1-r},
$$

we have 

$$
A^k\mathbf{x}_0 = \begin{pmatrix}1-2^k+3^k&1-3\cdot2^k+2\cdot3^k&-2^k+3^k\\2^k-3^k&3\cdot2^k-2\cdot3^k&2^k-3^k\\-1-2^k+2\cdot3^k&-1-3\cdot2^k+4\cdot3^k&-2^k+2\cdot3^k\end{pmatrix}
\begin{pmatrix}0\\-1\\2\end{pmatrix} = \begin{pmatrix}-1+2^k\\-2^k\\1+2^k\end{pmatrix},
$$

and

$$
\sum_{r=0}^{k-1}A^r\mathbf{f}_{k-1-r} &= \sum_{r=0}^{n-1} \begin{pmatrix}1-2^r+3^r&1-3\cdot2^r+2\cdot3^r&-2^r+3^r\\2^r-3^r&3\cdot2^r-2\cdot3^r&2^r-3^r\\-1-2^r+2\cdot3^r&-1-3\cdot2^r+4\cdot3^r&-2^r+2\cdot3^r\end{pmatrix}
\begin{pmatrix}1\\0\\0\end{pmatrix},\\
&=\sum_{r=0}^{k-1}\begin{pmatrix}1-2^r+3^r\\2^r-3^r\\-1-2^r+2\cdot3^r\end{pmatrix}.
$$

Now, 

$$
\sum_{r=0}^{k-1}1=k,\sum_{r=0}^{k-1}2^r=2^k-1 \text{ and } \sum_{r=0}^{k-1}3^r=\frac{3^k-1}{2}.
$$

Therefore

$$
\mathbf{x}_k &= \begin{pmatrix}-1+2^k\\-2^k\\1+2^k\end{pmatrix} + 
\begin{pmatrix} k-(2^k-1)+\frac{3^k-1}{2}\\2^k-1-\frac{3^k-1}{2}\\-k-(2^k-1)+(3^k-1)\end{pmatrix},\\
&=\begin{pmatrix} -\frac{1}{2}+k+\frac{3^k}{2},&-\frac{1}{2}-\frac{3^k}{2},&1-k+3^k\end{pmatrix}^T.
$$
